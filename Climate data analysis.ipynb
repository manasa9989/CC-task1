```python
# Replace 's' with NaN
daily_data.replace('s', pd.NA, inplace=True)
hourly_data.replace('s', pd.NA, inplace=True)
monthly_data.replace('s', pd.NA, inplace=True)
three_hour_data.replace('s', pd.NA, inplace=True)
```

```python
daily_data.head()
```

```python
# Handle missing values
daily_data.dropna(inplace=True)
hourly_data.dropna(inplace=True)
monthly_data.dropna(inplace=True)
three_hour_data.dropna(inplace=True)
```

```python
def replace_t(x):
    return np.nan if x == 'T' else x

# Apply the function to each DataFrame
daily_data = daily_data.applymap(replace_t)
hourly_data = hourly_data.applymap(replace_t)
monthly_data = monthly_data.applymap(replace_t)
three_hour_data = three_hour_data.applymap(replace_t)
```

```python
import matplotlib.pyplot as plt
precipitation_counts = daily_data['DailyPrecipitation'].value_counts()
plt.figure(figsize=(10, 8))

# Plot pie chart
labels = [f'{label} ({count})' for label, count in precipitation_counts.items()]  
plt.pie(precipitation_counts, labels=labels, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 10})  
plt.title('Distribution of Precipitation ')
plt.axis('equal')  
plt.legend(loc='upper right')  

# Show the plot
plt.show()
```

```python
#DAILY SNOWFALL
import matplotlib.pyplot as plt
snowfall_counts = daily_data['DailySnowfall'].value_counts()

plt.figure(figsize=(10, 8))

# Plot pie chart
labels = [f'{label} ({count})' for label, count in snowfall_counts.items()]  
plt.pie(snowfall_counts, labels=labels, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 10})  

plt.title('Distribution of Daily Snowfall Types')
plt.axis('equal')  
plt.legend(loc='upper right')  

# Show the plot
plt.show()
```
